### K fold prototype

```python
from sklearn.datasets import load_digits
mnist=load_digits(n_class=10)
```

```python
data=mnist.data
data.shape
```

```python
label=mnist.target
label.shape
```

```python
from sklearn.model_selection import KFold

seed=10
kfold=KFold(n_splits=2, shuffle=False)
```

```python
for train_index, test_index in kfold.split(data):
  print("TRAIN INDEX ", train_index, " TEST INDEX ", test_index)
  x_train, x_test=data[train_index], data[test_index]
  y_train, y_test=label[train_index], label[test_index]
  print("X train shape =", x_train.shape, "X test shape = ", x_test.shape)
```

```python
from torch.utils.data import DataLoader
train_dataloader=DataLoader(data, sampler=train_index, batch_size=2)
```

```python
js=pd.read_json("bert_base_output/eval_dynamics/dynamics_epoch_0.jsonl", orient="records", lines=True)
js
```

```python
js.sort_values(by="guid", inplace=True)
js
```

```python
results=pd.read_json("roberta_base_output/eval_dynamics/dynamics_epoch_3.jsonl", orient="records", lines=True)
results
```

```python
preds=[]
for confidences in results["logits_epoch_3"]:
  preds.append(np.argmax(confidences))
```

```python
results["prediction"]=preds
results
```

```python
orig=pd.read_csv("/content/gdrive/MyDrive/Maryland Cartography/Data/Winogrande/train.tsv", sep="\t")
incorrect_guids=[]
for i in range(len(results)):
  sample=results.iloc[i,:]
  if sample["gold"] != sample["prediction"]:
    incorrect_guids.append(sample["guid"])    
```

```python
for i in incorrect_guids[:100]:
  a=orig[orig.guid == i]
  b=results[results.guid == i]
  print(a["sentence"].item(),"\n Options", a["option1"].item(), " ", a["option2"].item(),"\n Gold: ", a["label"].item()," Pred: ", b["prediction"].item())
```
